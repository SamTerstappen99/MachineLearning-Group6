```python
import pandas as pd
data =  pd.read_csv("olist_data.csv")
```

```python
gemiddelde_locatie = data.groupby('geolocation_city')[['geolocation_lat', 'geolocation_lng']].mean()
bedrag_per_stad = data.groupby('geolocation_city')['price'].sum()
gemiddelde_locatie['total_price'] = bedrag_per_stad
gemiddelde_locatie = gemiddelde_locatie.reset_index()

# Toon de resultaten
gemiddelde_locatie
```

```python
from sklearn.cluster import KMeans
import numpy as np

geolocaties = gemiddelde_locatie[['geolocation_lat', 'geolocation_lng']].values

aantal_clusters = 100

# Voer K-means clustering uit
kmeans = KMeans(n_clusters=aantal_clusters, random_state=0)
kmeans.fit(geolocaties)

# Voeg de toegewezen clusterlabels toe aan de DataFrame
gemiddelde_locatie['cluster_label'] = kmeans.labels_
# Set de clusterlabels als index


# Toon de resultaten
gemiddelde_locatie
```

```python
# Groepeer de data op basis van de clusterlabels
grouped = gemiddelde_locatie.groupby('cluster_label')

# Voer verschillende aggregatiefuncties uit op verschillende kolommen
cluster_stats = grouped.agg({
    'geolocation_lat': 'mean',  # Gemiddelde van geolocation_lat
    'geolocation_lng': 'mean',  # Gemiddelde van geolocation_lng
    'total_price': 'sum'               # Som van prijs
})

# Toon de resultaten
print(cluster_stats)

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Maak een nieuwe figuur en 3D-as
fig = plt.figure(figsize=(120, 8))
ax = fig.add_subplot(111, projection='3d')

# Loop over elk cluster en teken een staaf op het XY-vlak met hoogte gelijk aan total_price
for index, row in cluster_stats.iterrows():
    ax.bar3d(row['geolocation_lng'], row['geolocation_lat'], 0, 1, 1, row['total_price'], color='skyblue')

# Stel de assenlabels in
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('Total Price')

# Titel van de plot
plt.title('Total Price per Cluster in XY Plane')

# Toon de plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Maak een nieuwe figuur en 3D-as
fig = plt.figure(figsize=(120, 8))
ax = fig.add_subplot(111, projection='3d')

# Definieer kleurverloop (groen) van lage hoogte (blauw) naar hoge hoogte (groen)
norm = plt.Normalize(cluster_stats['total_price'].min(), cluster_stats['total_price'].max())
cmap = plt.cm.Greens

# Loop over elk cluster en teken een staaf op het XY-vlak met hoogte gelijk aan total_price
for index, row in cluster_stats.iterrows():
    ax.bar3d(row['geolocation_lng'], row['geolocation_lat'], 0, 1, 1, row['total_price'], color=cmap(norm(row['total_price'])))

# Stel de assenlabels in
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('Total Price')

# Maak een kleurenbalk voor referentie
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, pad=0.05, label='Total Price')

# Titel van de plot
plt.title('Total Price per Cluster in XY Plane')

# Toon de plot
plt.show()

```

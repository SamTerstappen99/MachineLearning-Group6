```python
import pandas as pd
```

```python
customer = pd.read_csv("olist_customers_dataset.csv")
print(len(customer))
customer.head()
```

```python
geo = pd.read_csv("olist_geolocation_dataset.csv")
geo.drop_duplicates(subset="geolocation_zip_code_prefix", keep =  "last", inplace= True)
print(len(geo))
geo.head()
```

```python
cus_geo = customer.merge(geo, "left", left_on=["customer_zip_code_prefix"], right_on=["geolocation_zip_code_prefix"])
print(len(cus_geo))
cus_geo.head()
```

```python
order_p = pd.read_csv("olist_order_items_dataset.csv")
print(len(order_p))
order_p.head()
```

```python
order_c = pd.read_csv("olist_orders_dataset.csv")
print(len(order_c))
order_c.head()
```

```python
order =  order_c.merge(order_p, "left", on = "order_id")
order.head()
```

```python
prod = pd.read_csv("olist_products_dataset.csv")
prod_name_en = pd.read_csv("product_category_name_translation.csv")
print(len(prod))

prod =  prod.merge(prod_name_en, "left", on= "product_category_name")

print(len(prod))
prod.head()
```

```python
cus_geo_order = order.merge(cus_geo, "left", on = "customer_id")
print(len(cus_geo_order))
cus_geo_order.head()
```

```python
cus_geo_order_prod = cus_geo_order.merge(prod, "left", on = "product_id")
print(len(cus_geo_order_prod))
cus_geo_order_prod.head()
```

```python
data = cus_geo_order_prod[["order_id","product_id", "price","product_category_name_english", "customer_id","geolocation_lat","geolocation_lng", "geolocation_city","geolocation_state"]]
print(len(data))
```

```python
data.to_csv("olist_data.csv")
```

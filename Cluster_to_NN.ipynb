```python
import pandas as pd
data =  pd.read_csv("olist_data.csv")
data
```

```python
data.fillna(0, inplace=True)
```

```python
from sklearn.cluster import KMeans
import numpy as np

geolocaties = data[['geolocation_lat', 'geolocation_lng']].values

aantal_clusters = 100

# Voer K-means clustering uit
kmeans = KMeans(n_clusters=aantal_clusters, random_state=0)
kmeans.fit(geolocaties)

# Voeg de toegewezen clusterlabels toe aan de DataFrame
data['cluster_label'] = kmeans.labels_

data
```

```python
# Zorg ervoor dat alle waarden in de kolom 'product_category_name_english' strings zijn
from sklearn.preprocessing import LabelEncoder


data['product_category_name_english'] = data['product_category_name_english'].astype(str)

# Voer de LabelEncoder opnieuw uit en wijs de resultaten toe aan het originele DataFrame
data.loc[:, 'product_category_name_english'] = LabelEncoder().fit_transform(data['product_category_name_english'])

# Controleer of de wijzigingen zijn toegepast
data

```

```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler


# Selecteer kenmerken en labels
X = data[['cluster_label','product_category_name_english']]  # Selecteer kenmerken
y = data['price']  # Selecteer label

# print("Shape of X:", X.shape)
# print("Shape of y:", y.shape)

# Normaliseer de kenmerken
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split de dataset in trainings- en testsets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Bouw het neuraal netwerkmodel
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compileer het model
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])

# Train het model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Evaluatie van het model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test loss:", loss)
print("Test accuracy:", accuracy)
```
